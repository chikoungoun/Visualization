# Styling with Pandas

Styling enchance the readability of the plots and make it more enjoyable and easier to decypher

```python
import pandas as pd
%matplotlib inline
reviews = pd.read_csv('Dataset/winemag-data_first150k.csv',index_col=0)
```

```python
reviews.head()
```

## Points on style

Using various options to make the plots more efficients

### * make the bars bigger : Figsize

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12,6))
```

figsize= (width,height) : tuple for width and height (largeur et longueur)

### * color : change the color of the bars

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12,6),color='purple')
```

changing the color of the bars with :
color='color_name'

### * Fontsize

```python
reviews['points'].value_counts().sort_index().plot.bar(fontsize=20,color='green',figsize=(12,6))
```

### * Title 

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12,6),color='red',fontsize=19,title='Rankings Given by Wine Magazine')
```

title = 'Text'

**-> It's not possible to change the size of the title with pandas only. We'll need to use the matplotlib options that pandas is based on; but gives more options when called apart**

```python
import matplotlib.pyplot as plt
ax = reviews['points'].value_counts().sort_index().plot.bar(figsize=(12,6),
                                                          color='mediumvioletred',
                                                          fontsize=19)
ax.set_title("Rankings Given bu Wine Magazine",fontsize=22)
```
