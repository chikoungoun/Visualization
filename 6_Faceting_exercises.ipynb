# Faceting with Pokemon DB

```python
import pandas as pd
import seaborn as sns
```

```python
pokemon = pd.read_csv('Dataset/Pokemon_new.csv',index_col=0)
pokemon.head(3)
```

**Pokemon Attack distribution within legendary pokemons or not : **

```python
g = sns.FacetGrid(pokemon, row='Legendary')
g.map(sns.kdeplot,'Attack')
```

** Pokemon Attack distribution by legedary and generation : **

```python
g = sns.FacetGrid(pokemon, row='Generation',col='Legendary')
g.map(sns.kdeplot,'Attack')
```

** Pairplot for HP, Attack and defense : **

```python
sns.pairplot(pokemon[['HP','Attack','Defense']])
```
