# <u>FancyArrow Atelier</u>

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.patches as patches
```

```python
x = 0.1
y = 0.5
end_x = 0.9
end_y = 0.8
dx = end_x - x
dy = end_y - y
```

## Arrow

To draw with "Arrow", we need to use the **x,y,dx,dy** arguments.

"Arrow" doesn't give a pretty enough result. Arrows seem distorted

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

#Drawing an arrow with the "Arrow" Object
# uses x,y, dx and dy
arrow = patches.Arrow(x,y,dx,dy)
ax[0].add_patch(arrow)

#width to control the arrow width
arrow = patches.Arrow(x,y-.4,dx,dy,width=.5,color="C3")
ax[1].add_patch(arrow)
```

## FancyArrow

PLaying with the different parameters

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))


# Drawing with FancyArrow
arrow = patches.FancyArrow(x,y,dx,dy,head_starts_at_zero=True,overhang=0.5,width=.01,head_width=.1,head_length=.1,shape='full',color="C1")
ax[0].add_patch(arrow)
```

### FancyArrow properties 

**color** property overrides the edge and face color

**edgecolor** and **facecolor** work with a (R,G,B,alpha) system
* Facecolor : fills the arrow color
* Edgecolor : fills the borders of the arrow

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

"""
alpha : opacity level
color : arrow color


"""
# Drawing with FancyArrow
arrow = patches.FancyArrow(x,y,dx,dy,width=.05,head_width=0,head_length=0,edgecolor=(1,0,0,0.1),facecolor=(1,1,0,1))
ax[0].add_patch(arrow)
```

## FancyArrowPatch

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

"""
creating the arrow
mutation_scale : size of the arrow drawing
"""
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50)

#adding the arrow on the plot
ax[0].add_patch(arrow)

arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50)
ax[1].add_patch(arrow)
#change the proportions x and y coordinates
ax[1].set(xlim=(0,2),ylim=(0,2))
```

## FancyArrowPatch : Keeping the arrow size even if the proportions change

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

"""
creating the arrow
mutation_scale : size of the arrow drawing
"""
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50,transform=ax[0].transAxes)

#adding the arrow on the plot
ax[0].add_patch(arrow)

arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50,transform=ax[1].transAxes)

ax[1].add_patch(arrow)
ax[1].set(xlim=(0,2),ylim=(0,2))
```

### FancyArrowPatch properties : ArrowStyle

source : https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.patches.ArrowStyle.html#matplotlib.patches.ArrowStyle

```python
from matplotlib.patches import ArrowStyle
```

```python
fig,ax =  plt.subplots(4,2,figsize=(10,10))

#1. Normal Arrow
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50,transform=ax[0][0].transAxes)
ax[0][0].add_patch(arrow)
ax[0][0].set_title("yo")

#2. add a tail to the rear
AS = ArrowStyle("Fancy", head_length=.4, head_width=.4, tail_width=.4)
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=50,transform=ax[0][1].transAxes)
ax[0][1].add_patch(arrow)
ax[0][1].set(xlim=(0,2),ylim=(0,2))
ax[0][1].set_title("Tail with 'Fancy'")

#3. "<->"
AS = ArrowStyle("<->")
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=50,transform=ax[1][0].transAxes,color="orange")
ax[1][0].add_patch(arrow)
ax[1][0].set_title("<->")

#4. "<|-|>"
AS = ArrowStyle("<|-|>")
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=50,color="red")
ax[1][1].add_patch(arrow)
ax[1][1].set_title("<|-|>")

#5. "<|-|>"
AS = ArrowStyle("wedge")
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=50,color="red")
ax[2][0].add_patch(arrow)
ax[2][0].set_title("wedge")

#5. "]-["
AS = ArrowStyle("]-[")
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=10,color="red")
ax[2][1].add_patch(arrow)
ax[2][1].set_title("]-[")


#5. "|-|"
AS = ArrowStyle("|-|")
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=10,color="red")
ax[3][0].add_patch(arrow)
ax[3][0].set_title("|-|")

```

**ArrowStyle can be called in another way**  
here we instanciated the "<->" using : ArrowStyle.CurveAB()

```python
fig,ax =  plt.subplots(2,2,figsize=(10,10))

#2. add a tail to the rear
AS = ArrowStyle.CurveAB()
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),arrowstyle=AS ,mutation_scale=50,transform=ax[0][1].transAxes)
ax[0][1].add_patch(arrow)
ax[0][1].set(xlim=(0,2),ylim=(0,2))
ax[0][1].set_title("Tail with 'Fancy'")
```
