# <u>FancyArrow Atelier</u>

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.patches as patches
```

```python
x = 0.1
y = 0.5
end_x = 0.9
end_y = 0.8
dx = end_x - x
dy = end_y - y
```

### FancyArrowPatch

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

"""
creating the arrow
mutation_scale : size of the arrow drawing
"""
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50)

#adding the arrow on the plot
ax[0].add_patch(arrow)

arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50)
ax[1].add_patch(arrow)
#change the proportions x and y coordinates
ax[1].set(xlim=(0,2),ylim=(0,2))
```

### FancyArrowPatch : Keeping the arrow size even if the proportions change

```python
# creating the subplots 
fig,ax =  plt.subplots(2,figsize=(5,5))

"""
creating the arrow
mutation_scale : size of the arrow drawing
"""
arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50,transform=ax[0].transAxes)

#adding the arrow on the plot
ax[0].add_patch(arrow)

arrow = patches.FancyArrowPatch((x,y),(end_x,end_y),mutation_scale=50,transform=ax[1].transAxes)

ax[1].add_patch(arrow)
ax[1].set(xlim=(0,2),ylim=(0,2))
```
