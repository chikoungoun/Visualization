```python
import pandas as pd
reviews = pd.read_csv("Dataset/winemag-data_first150k.csv", index_col=0)
reviews.head()
```

## Scatter Plot

2-dimensional data visualization representing values obtained with the help of 2 variables (x and y).
* Relationship between 2 variables
* also called correlation plots

```python
#prendre un plus petit chatillon de donnes pour viter la confusion et la superposition des donnes
reviews[reviews['price'] < 100].sample(100).plot.scatter(x='price',y='points')
```

```python
#sans chantillonage
reviews[reviews['price'] < 100].plot.scatter(x='price',y='points')
```

# Hexplot

Close to a scatter plot, shows a sort of heat map of the data and is useful in case of **overlapping**

```python
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15)
```

## Stacked Plot

Helpful with categorical data
* Part to whole relationship
* used if the total of parts is crucial for better comparison
* They work best for nominal categorical or small ordinal categorical variables.
* (-) not very precise for interpretability
* (-) the number of values should be limited for better readability

```python
wine_counts = pd.read_csv('Dataset/top-five-wine-score-counts.csv',index_col=0)
```

```python
wine_counts
```

Comparison of points repartition for the 5 different wines :

```python
wine_counts.plot.bar(stacked=True)
```

Area Chart version :

```python
wine_counts.plot.area()
```

## Bivariate line chart

much more interpretable than the stacked plot version

```python
wine_counts.plot.line()
```
