```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# Plotting Seaborn with Pokemon data

```python
pokemon = pd.read_csv('Dataset/Pokemon_new.csv',index_col=0)
pokemon.head()
```

```python
pokemon.columns
```

**Generation count : ** (barchart)

```python
sns.countplot(pokemon['Generation'])
```

** HP : ** (histogram and kde)

```python
sns.distplot(pokemon['HP'])
```

** Attack / Defense : ** (Scatterplot and Hexplot)

```python
sns.jointplot(data=pokemon,x='Attack',y='Defense')
```

```python
sns.jointplot(data=pokemon,x='Attack',y='Defense',kind='hex')
```

** HP / Attack : **(KDE circles)

```python
sns.kdeplot(pokemon['Attack'],pokemon['Defense'])
```

** Legendary / Attack : **

Boxplot

```python
sns.boxplot(pokemon['Legendary'],pokemon['Attack'])
```

violin plot version : 

```python
sns.violinplot(pokemon['Legendary'],pokemon['Attack'])
```
