```python
import pandas as pd
reviews  = pd.read_csv('Dataset/winemag-data_first150k.csv',index_col=0)
reviews.head()
```

creating multiple plots side by side in one big figure.

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axarr = plt.subplots(2,1, figsize=(12,8))
```

** Pass a plot into a subplot **

```python
fig, axarr = plt.subplots(2,1, figsize=(12,8))

#first plot
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0])

#second plot
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1])
```

**Making the call different with many columns : **

```python
fig, axarr = plt.subplots(2,2,figsize=(12,8))

reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0])

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1])
```

** Plotting with style : **

```python
fig, axarr = plt.subplots(2,2,figsize=(12,8))

#First Plot
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0],
                                                      fontsize=12,
                                                      color='mediumvioletred')

axarr[0][0].set_title("Wine Scores", fontsize=18)

#Second plot
reviews['variety'].value_counts().head(20).plot.bar(ax=axarr[1][0],
                                                   fontsize=12,
                                                   color='green')
axarr[1][0].set_title('Wine Varieties',fontsize=18)

#Third plot
reviews['price'].value_counts().plot.hist(
    ax=axarr[0][1], fontsize=12, color='red'
)
axarr[0][1].set_title("Wine Prices", fontsize=18)



#Fourth plot
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1],
                                                   fontsize=12,
                                                   color='blue')
axarr[1][1].set_title("Wine Origins", fontsize=18)


plt.subplots_adjust(hspace=.3)#space between subplots

import seaborn as sns
sns.despine()
```
