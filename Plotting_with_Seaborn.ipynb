# Plotting With Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. - Seaborn definition

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
reviews = pd.read_csv("Dataset/winemag-data_first150k.csv",index_col=0)
```

## Countplot

aggregates the data as like of applying '.value_counts()' and represents categorical data like a bar chart with colorful bins

```python
sns.countplot(reviews['points'])
```

## KDE Plot

Kernel density estimation is a fundamental data smoothing problem. 
* Works well for interval data
* (-) not so much for categorical data

**with line chart :** 

```python
reviews[reviews['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

**with KDE Plot :**

```python
sns.kdeplot(reviews.query('price < 200').price)
```

KDE plot is better than a line chart for getting the "true shape" of interval data. 

## Distplot

equivalent to the pandas 'histogram'

**Pandas Version : **

```python
reviews['points'].plot.hist(bins=10)
```

**Seaborn Version :**

```python
sns.distplot(reviews['points'],bins=10,kde=False)
```

## Jointplot (Scatterplot x Hexplot)


equivalent to pandas Scatteplot and hexplot

**Scatterplot : **

```python
sns.jointplot(x='price',y='points', data=reviews[reviews['price'] < 100])
```

**Hexplot :**

```python
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 100], kind='hex', 
              gridsize=20)
```

## Boxplot and Violin plot

```python
df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]
```

**Boxplot :**

```python
sns.boxplot(
    x='variety',
    y='points',
    data=df
)
```

**Violin plot : **

```python
sns.violinplot(
    x='variety',
    y='points',
    data=reviews[reviews.variety.isin(reviews.variety.value_counts()[:5].index)]
)
```
