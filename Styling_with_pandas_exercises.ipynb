# Plot Styling exercises with the pokemon data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv('Dataset/Pokemon_new.csv',index_col=0)
pokemon.head(3)
```

```python
pokemon.columns
```

**Pokemon Attack - Defense relationship**

```python
pokemon.plot.scatter(x='Attack',y='Defense',figsize=(12,6),title='Pokemon by Attack and Defense')
```

**Pokemon by Stat Total** : (bins : number of bins)

```python
ax = pokemon['Total'].plot.hist(figsize=(12,6),color='grey',bins=50,fontsize=14)
ax.set_title('Pokemon by stat Total',fontsize=20)
```

**Pokemon by primary Type**

```python
ax = pokemon['Type 1'].value_counts().plot.bar(figsize=(12,6),fontsize=14)
ax.set_title('Pokemon by Primary Type')
sns.despine(bottom=True,left=True)
```
